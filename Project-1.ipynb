```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Debt data files
Total_Debt_path = "data/Total_Debt_2007_2022.csv"

# Read the Debt data
Total_Debt = pd.read_csv(Total_Debt_path)

# display
Total_Debt
```

```python
# Setting the size for the figure
plt.figure(figsize=(12, 6))

# Width of a bar 
width = 0.4

# Creating a bar graph for 'Total Dollars'
bars1 = plt.bar(Total_Debt['Federal Fiscal Year'] - width/2, Total_Debt['Total Dollars ($B)'], width=width, label='Total Dollars ($B)', color='green', alpha=0.6)

# # Creating a secondary y-axis for 'Total Recipients'
ax2 = plt.gca().twinx()
bars2 = ax2.bar(Total_Debt['Federal Fiscal Year'] + width/2, Total_Debt['Total Recipients (M)'], width=width, label='Total Recipients (M)', color='blue', alpha=0.6)

# Setting the x and y axis labels
plt.xlabel('Federal Fiscal Year')
plt.ylabel('Total Dollars ($B)')
plt.ylim(0, 200)
ax2.set_ylabel('Total Recipients (M)')

# Setting the title for the graph
plt.title('Total Dollars and Total Recipients Over the Years (DL, FFEL, Perkin)')

# Displaying the legend
plt.legend(loc='upper left', handles=[bars1, bars2])
ax2.legend(loc='upper left')

# Displaying the graph
plt.tight_layout()
plt.show()
```

```python
# Setting the size for the figure
plt.figure(figsize=(12, 6))

# Plotting the line graphs for recipients
plt.plot(Total_Debt['Federal Fiscal Year'], Total_Debt['Recipients (M)'], marker='o', label='Direct Loans Recipients (M)', color='green')
plt.plot(Total_Debt['Federal Fiscal Year'], Total_Debt['FFEL Recipients (M)'], marker='o', label='FFEL Recipients (M)', color='blue')
plt.plot(Total_Debt['Federal Fiscal Year'], Total_Debt['Perkins Recipients (M)'], marker='o', label='Perkins Recipients (M)', color='red')

# Setting the x and y axis labels
plt.xlabel('Federal Fiscal Year')
plt.ylabel('Recipients (Millions)')

# Setting the title for the graph
plt.title('Loan Recipients Over the Years')

# Displaying the legend
plt.legend()

# Displaying the graph
plt.tight_layout()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
Total_Debt['Direct Loans ($B)'] = Total_Debt['Direct Loans ($B)'].astype(float)
Total_Debt['FFEL ($B)'] = Total_Debt['FFEL ($B)'].astype(float)
Total_Debt['Perkins Loans ($B)'] = Total_Debt['Perkins Loans ($B)'].astype(float)
```

```python

sum_direct_loans = Total_Debt['Direct Loans ($B)'].sum()
sum_FFEL = Total_Debt['FFEL ($B)'].sum()
sum_Perkins = Total_Debt['Perkins Loans ($B)'].sum()

# Data to plot
labels = 'Direct Loans', 'FFEL', 'Perkins Loans'
sizes = [sum_direct_loans, sum_FFEL, sum_Perkins]
colors = ['yellow', 'green', 'red']
explode = (0.1, 0, 0)  # explode 1st slice for emphasis

# Plot
plt.figure(figsize=(10,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Distribution of Direct Loans, FFEL, and Perkins Loans over the years')
plt.show()
```
