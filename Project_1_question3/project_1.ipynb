```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

#Data files
tuition_data_path = "project_1/tuition.csv"
debt_borrower_path = "project_1/debt_borrower.csv"
```

```python
tuition_data = pd.read_csv('project_1/tuition.csv',header=[0,1,2],index_col=[0],engine='python')

for column in tuition_data.columns:
    tuition_data[column] = tuition_data[column].str.replace(',','').astype(float)

tuition_data.head()
```

```python
#Create line graph for tuition costs for all 4 year universities 

x = tuition_data.index.str.split("-").str[0].astype(int)
y = tuition_data[('Constant 2021-22 dollars','Total tuition, fees, room, and board','4-year')]

#label x & Y axis 
plt.xlabel("Year")
plt.ylabel("Total tuition, fees, room, and board ($)")
plt.title("Tuition costs from 1963-2022 for 4-year institutions")

#Plot graph
plt.plot(x,y)
plt.show
```

```python
debt_borrower = pd.read_csv('project_1/debt_borrower.csv', skiprows=[3],header=[3,4],index_col=[0,1],skipfooter=2,engine='python')

debt_borrower.head(30)
```

```python
#plotfor debt 20 to 40 K


x = debt_borrower.index
y = debt_borrower.loc[:,('20K to 40K', 'Dollars Outstanding            (in billions)')]
x2 = [f"{n[0]}-{n[1]}" for n in x]
print(x2)


#label x & Y axis 
plt.xlabel("Year")
plt.ylabel("Dollars Outstanding (in billions)")
plt.title("Year vs Debt (20K to 40K)")

#Plot graph
plt.plot(x2,y)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

```python
# Plot borrower data

x = debt_borrower.index
y1 = debt_borrower.loc[:,('20K to 40K', 'Borrowers     (in millions)')]
x2 = [f"{n[0]}-{n[1]}" for n in x]


#label x & Y axis 
plt.xlabel("Year")
plt.ylabel("Borrowers  (in millions)")
plt.title("Year vs Borrower 20K to 40K")

#Plot graph
plt.plot(x2,y1, color='green')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```
